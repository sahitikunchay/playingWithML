# Ad Detection

```python
import pandas as pd
import numpy as np
```

```python
dataFile='C:/Users/sahiti_kunchay/Desktop/ad-dataset/ad.data'
data=pd.read_csv(dataFile,sep=",",header=None,low_memory=False)
```

```python
data.head(5)
```

```python
#checking for missing values and replacing it with NaN
def toNum(cell):
    try:
        return np.float(cell)
    except:
        return np.nan
    
def seriesToNum(series):
    return series.apply(toNum)
```

```python
trainData = data.iloc[0:, 0:-1].apply(seriesToNum)
trainData.head(11)
```

```python
trainData = trainData.dropna()
trainData.head(11)
```

```python
def toLabel(str):
    if(str)=='ad.':
        return 1
    else:
        return 0
    
trainLabels=data.iloc[trainData.index, -1].apply(toLabel)
trainLabels.head
```

```python
#Training
from sklearn.svm import LinearSVC

classifier = LinearSVC()
classifier.fit(trainData[100:2300], trainLabels[100:2300])
```

```python
#Testing
classifier.predict(trainData.iloc[12].reshape(1, -1))
```
