```python
train_documents = [line[0] for line in lines ]
train_documents
```

```python
train_labels = [line[1] for line in lines ]
train_labels
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
countVectorizer = CountVectorizer(binary='true')
train_documents = countVectorizer.fit_transform(train_documents)
```

```python
train_documents
```

```python
print(train_documents[0])
```

```python
from sklearn.naive_bayes import BernoulliNB
classifier = BernoulliNB().fit(train_documents, train_labels)
```

```python
#testing
classifier.predict(countVectorizer.transform(["i kinda dont like this movie"]))
```
